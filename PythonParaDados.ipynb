```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('saudedosono.csv')
df.head()
```

```python
#1) Ao visualizar a base você percebeu que seria melhor alterar o nome de
#algumas colunas. Mude o ‘ID’ para ‘Identificador’, corrija o nome da
#coluna que indica a pressão sanguínea, mude a coluna ‘Ocupação’ para
#‘Profissão’, a coluna ‘Categoria BMI’ está em parte em inglês, substitua
#para ‘Categoria IMC’.

colsnew = {'ID': 'Identificador',
           'Pressão sanguíneaaaa': 'Pressão sanguínea',
           'Ocupação': 'Profissão',
           'Categoria BMI': 'Categoria IMC'
}

# Substituindo os nomes das colunas
df = df.rename(columns=colsnew)

df.head()
```

```python
#2) Qual é a média, a moda e a mediana de horas de sono para cada uma
#das profissões? [‘mean’, np.median, pd.Series.mod]

print('Média:')
dfmedia = df[['Profissão', 'Duração do sono']].groupby('Profissão')['Duração do sono'].mean()
print(dfmedia)
print('Moda:')
dfmoda = df[['Profissão', 'Duração do sono']].groupby('Profissão')['Duração do sono'].agg(pd.Series.mode)
print(dfmoda)
print('Mediana:')
dfmediana = df[['Profissão', 'Duração do sono']].groupby('Profissão')['Duração do sono'].agg(np.median)
print(dfmediana)
```

```python
#3)  Das pessoas que atuam com engenharia de software qual a
#porcentagem de obesos?

percobs = (df['Profissão'] == 'Eng. de Software') & (df['Categoria IMC'] == 'Obesidade')
perengobs = (percobs.sum() / (df['Profissão'] == 'Eng. de Software').sum()) * 100
print(perengobs)
```

```python
#4 De acordo com os dados, advogar ou ser representante de vendas faz
#você dormir menos? (Use o método ‘isin’, considere a média)
profissoes = ['Advogado(a)', 'Representante de Vendas']

media = df[df['Profissão'].isin(profissoes)]['Duração do sono'].mean()
print('A média de sono de advogados e representante de vendas é: ', media)
```

```python
# 5) Entre quem fez enfermagem e quem fez medicina, quem tem menos
#horas de sono? (Use o método ‘isin’, considere a média)
mediaenf = df[df['Profissão'].isin(['Enfermeiro(a)'])]['Duração do sono'].mean()
mediamed = df[df['Profissão'].isin(['Médico(a)'])]['Duração do sono'].mean()
print(f'A menor média de sono é para médicos: {mediamed:.2f} horas, enquanto enfermeiros possuem: {mediaenf:.2f}')
```

```python
#6) Faça um subconjunto com as colunas Identificador, Gênero, Idade,
#Pressão sanguínea e Frequência cardíaca.
dfsub = df[['Identificador', 'Gênero', 'Idade', 'Pressão sanguínea', 'Frequência cardíaca']]
dfsub.head()
```

```python
#7) Descubra qual a profissão menos frequente no conjunto. (Use value_counts)

print('A profissão menos frequente é: ',df['Profissão'].value_counts().idxmin())
```

```python
# 8) Quem tem maior pressão sanguínea média, homens ou mulheres? (Considere a média)
dfsub = dfsub.copy()
dfsub['Pressão sanguínea'] = dfsub['Pressão sanguínea'].apply(lambda x: (int(x.split('/')[0]) + int(x.split('/')[1])) / 2)
print(dfsub.groupby('Gênero')['Pressão sanguínea'].mean())
print('Mulheres possuem maior pressão sanguínea')
```

```python
#9) É predominante entre os participantes dormir 8 horas por dia (considere usar Moda como medida)?

print('É predominante dormir em torno de: ',df['Duração do sono'].agg(pd.Series.mode))
```

```python
#10) Pessoas com frequências cardíacas acima de 70 dão mais passos que pessoas com frequência cardíaca menor ou igual a 70? (Use a média)

maior = df[df['Frequência cardíaca'] > 70]['Passos diários'].mean()
menor = df[df['Frequência cardíaca'] <= 70]['Passos diários'].mean()

if maior > menor:
    print('Pessoas com frequências cardíacas acima de 70 dão mais passos.')
else:
    print('Pessoas com frequências cardíacas igual ou abaixo de 70 dão mais passos ou a média é igual.')
```
